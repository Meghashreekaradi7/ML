```python
import pandas as pd
import numpy as np

from sklearn.preprocessing import MinMaxScaler, StandardScaler, LabelEncoder
```

```python
import pandas as pd

diabetes_df = pd.read_csv("Dataset of Diabetes .csv")
diabetes_df.head()
```

```python
adult_df = pd.read_csv("adult (1).csv")
adult_df.head()
```

```python
for col in diabetes_df.select_dtypes(include=np.number).columns:
    diabetes_df[col] = diabetes_df[col].fillna(diabetes_df[col].mean())

for col in adult_df.select_dtypes(include="object").columns:
    adult_df[col] = adult_df[col].fillna(adult_df[col].mode()[0])
```

```python
from sklearn.preprocessing import LabelEncoder

encoder = LabelEncoder()
cat_cols = ["gender", "class"]

for col in cat_cols:
    diabetes_df[col] = encoder.fit_transform(diabetes_df[col])

print("Categorical Data Encoded")
```

```python
def remove_outliers(data, col):
    Q1 = data[col].quantile(0.25)
    Q3 = data[col].quantile(0.75)
    IQR = Q3 - Q1

    lower = Q1 - 1.5 * IQR
    upper = Q3 + 1.5 * IQR

    return data[(data[col] >= lower) & (data[col] <= upper)]

for col in diabetes_df.select_dtypes(include=np.number):
    diabetes_df = remove_outliers(diabetes_df, col)

print("Outliers Removed")
```

```python
from sklearn.preprocessing import MinMaxScaler

minmax = MinMaxScaler()

df_minmax = diabetes_df.copy()

features = df_minmax.drop("class", axis=1)

df_minmax[features.columns] = minmax.fit_transform(features)

print("Min-Max Normalized Data:")
df_minmax.head()
```

```python
from sklearn.preprocessing import StandardScaler

standard = StandardScaler()

df_standard = diabetes_df.copy()

features = df_standard.drop("class", axis=1)

df_standard[features.columns] = standard.fit_transform(features)

print("Standard Scaled Data:")
df_standard.head()
```
